```python
data = pd.read_csv('data_stocks.csv')
```

```python
data.head(10)
```

```python
data.set_index('DATE',inplace=True)
#help(data.reset_index)
```

```python
 data.head()
```

```python
from sklearn import cluster, datasets as dt
```

```python
n_clusters = 5
```

```python
np.random.seed(0)
k_Means = cluster.KMeans(n_clusters=n_clusters, n_init=4, max_iter=200)
k_Means.fit(data)
```

```python
k_Means.labels_
```

```python
len(k_Means.labels_)
```

```python
k_Means.cluster_centers_
```

```python
data['Cluster_No'] = k_Means.labels_
```

```python
companyMean = data.apply(np.mean, axis=0)
type(companyMean)
```

```python
companyStd = data.apply(np.std, axis=0)
type(companyStd)
```

```python
companyName = data.columns.tolist()[:len(data.columns)-1]
```

```python
companyData = pd.DataFrame({'Company': companyName, 'cMean': companyMean.tolist()[:len(data.columns)-1], 'cStd': companyStd.tolist()[:len(data.columns)-1]})
companyData.head()
```

```python
companyData.sort_values('cStd',ascending=False)
```

```python
df1 = data.loc[data.Cluster_No == 1, :]
df2 = data.loc[data.Cluster_No == 2, :]
df3 = data.loc[data.Cluster_No == 3, :]
df4 = data.loc[data.Cluster_No == 4, :]
df5 = data.loc[data.Cluster_No == 5, :]
```

```python
df1Mean = df1.apply(np.mean, axis=0).tolist()[:len(df1.columns)-1]
df1Std = df1.apply(np.std, axis=0).tolist()[:len(df1.columns)-1]

companyDf1 = pd.DataFrame({'Company': companyName, 'cMean': df1Mean, 'cStd': df1Std})

df2Mean = df2.apply(np.mean, axis=0).tolist()[:len(df2.columns)-1]
df2Std = df2.apply(np.std, axis=0).tolist()[:len(df2.columns)-1]

companyDf2 = pd.DataFrame({'Company': companyName, 'cMean': df2Mean, 'cStd': df2Std})

df3Mean = df3.apply(np.mean, axis=0).tolist()[:len(df3.columns)-1]
df3Std = df3.apply(np.std, axis=0).tolist()[:len(df3.columns)-1]

companyDf3 = pd.DataFrame({'Company': companyName, 'cMean': df3Mean, 'cStd': df3Std})

df4Mean = df4.apply(np.mean, axis=0).tolist()[:len(df4.columns)-1]
df4Std = df4.apply(np.std, axis=0).tolist()[:len(df4.columns)-1]

companyDf4 = pd.DataFrame({'Company': companyName, 'cMean': df4Mean, 'cStd': df4Std})

df5Mean = df5.apply(np.mean, axis=0).tolist()[:len(df5.columns)-1]
df5Std = df5.apply(np.std, axis=0).tolist()[:len(df5.columns)-1]

companyDf5 = pd.DataFrame({'Company': companyName, 'cMean': df5Mean, 'cStd': df5Std})
```

```python
companyDf1.set_index('Company',inplace=True)
companyDf2.set_index('Company',inplace=True)
companyDf3.set_index('Company',inplace=True)
companyDf4.set_index('Company',inplace=True)
companyDf5.set_index('Company',inplace=True)
```

```python
companyData['Cluster_1_Mean'] = companyDf1.cMean
companyData['Cluster_2_Mean'] = companyDf2.cMean
companyData['Cluster_3_Mean'] = companyDf3.cMean
companyData['Cluster_4_Mean'] = companyDf4.cMean
companyData['Cluster_5_Mean'] = companyDf5.cMean

companyData['Cluster_1_Std'] = companyDf1.cStd
companyData['Cluster_2_Std'] = companyDf2.cStd
companyData['Cluster_3_Std'] = companyDf3.cStd
companyData['Cluster_4_Std'] = companyDf4.cStd
companyData['Cluster_5_Std'] = companyDf5.cStd
```
